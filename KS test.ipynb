```r
S = c(0.9, 1.3, 2.8, 7.5, 9.1, 12.3, 15.1, 28.1, 48.9, 66.5)
length(S)
```

```r
for(i in 1:10){
FnF = i/10-pnorm(sort(S)[i]) # "Fn = fonction empirique", "F = fonction théorique" 
    cat(FnF, ",") }
```

```r
res1=c(-0.7159399 ,-0.7031995 ,-0.6974449 ,-0.6 ,-0.5 ,-0.4 ,-0.3 ,-0.2 ,-0.1 ,0)
max(abs(res1))
```

```r
res2=c(0.8159399 ,0.8031995 ,0.7974449 ,0.7 ,0.6 ,0.5 ,0.4 ,0.3 ,0.2 ,0.1)
max(abs(res2))
```

```r
options(repr.plot.width=5, repr.plot.height=5)
plot(ecdf(S), main="Empirical cumulative distribution function", col="blue")
curve(pnorm(S),xname = "S",lwd=2,col="red", add=TRUE)
```

#### Critical distance (n=10, alpha=0.05) = 0.40925

#### D > Dc alors on rejette l’hypothèse de nullité. Les données ne suivent pas une loi normale

## KS test : 2 samples

```r
S1 = c(0.9, 1.3, 2.8, 7.5, 9.1, 12.3, 15.1, 28.1, 48.9, 66.5)
length(S)
```

```r
KS1 = ks.test(S1, "pnorm")
KS1
```

```r
S2 = c(1.9, 1, 4.8, 1.5, 6.1, 14.2, 13.7, 34.4, 44.2, 50.1)
length(S2)
```

```r
KS2 = ks.test(S2, "pnorm")
KS2
```

```r
options(repr.plot.width=5, repr.plot.height=5)
plot(ecdf(S1), main="Fonctions de répartition", col="blue")
plot(ecdf(S2), col="red", add=TRUE)
curve(pnorm(S),xname = "S",lwd=2,col="red", add=TRUE)
```

```r
KS3 = ks.test(S1, S2, alternative="two.sided")
KS3
```
